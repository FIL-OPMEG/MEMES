```python
# Import variables
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt

# Set seaborn plotting defaults
rc={'font.size': 15, 'axes.labelsize': 25, 
    'axes.titlesize': 25,'xtick.labelsize': 20, 'ytick.labelsize': 20}

sns.set(style="white",rc=rc, font="Arial")
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
```

```python
# Set a colourful colourscheme
sns.set_palette("muted")
```

```python
# Load data
df = pd.read_csv('sourcemodel_data.csv')
df.tail()
```

## Distance error (cm) from peak sourcespace voxel

```python
# Plot and Save
ax = sns.boxplot(data=df,y='MEMES_data',x='MEMES_variation',fliersize=0,linewidth=1)
ax = sns.swarmplot(data=df,y='MEMES_data',x='MEMES_variation',color=".3",size=5)
plt.ylim(-1,5)
plt.ylabel("Distance Error (cm)")
plt.xlabel("")
ax.set_xticklabels(['MEMES','Average','Scaled','No Face'])
#plt.xlabel("Autism Quotient")
plt.savefig('source_model_peak_error.png',bbox_inches='tight',dpi=200)
```

## Difference in power (%) from peak sourcespace voxel

```python
# Load data
df_pow = pd.read_csv('sourcemodel_data_pow.csv')
```

```python
# Plot and Save
ax = sns.boxplot(data=df_pow,y='MEMES_data',x='MEMES_variation',fliersize=0,linewidth=1)
ax = sns.swarmplot(data=df_pow,y='MEMES_data',x='MEMES_variation',color=".3",size=5)
plt.ylim(-1,2)
plt.ylabel("Power Difference (%)")
plt.xlabel("")
ax.set_xticklabels(['MEMES','Average','Scaled','No Face'])
#plt.xlabel("Autism Quotient")
plt.savefig('source_model_peak_pow_diff.png',bbox_inches='tight',dpi=200)
```
